```python
import numpy as np
datos=np.loadtxt("manzanas.csv",delimiter=",", usecols=np.arange(1,88,1))
datos=datos.T
datos
```

```python
fechas=datos[:,0]
precios=datos[:,1:]
```

```python
Moscu= precios[:,0]
x=fechas
y=0.5*x+78
plt.plot(fechas, Moscu)
plt.plot(x,y)
```

```python
n=np.size(Moscu)
y=Moscu
x=fechas
```

```python
a=(n*np.sum(x*y)-np.sum(x)*np.sum(y))/((n*np.sum(x**2))-(np.sum(x)**2))
b=(np.mean(y)-a*np.mean(x))
print(a)
print(b)
```

```python
y=a*x+b
np.linalg.norm(Moscu-y)
```

```python
plt.plot(fechas, Moscu)
plt.plot(x,y)
plt.plot(25,a*25+b,'*b')
plt.plot(75,a*75+b,'_r')
```

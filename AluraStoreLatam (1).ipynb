```python
acomodoT1=sorted(conteo_productos.values())
acomodoT2=sorted(conteo_productos2.values())
acomodoT3=sorted(conteo_productos3.values())
acomodoT4=sorted(conteo_productos4.values())

dicP1=dict(zip(productos,conteo_productos.values()))
dicP2=dict(zip(productos,conteo_productos2.values()))
dicP3=dict(zip(productos,conteo_productos3.values()))
dicP4=dict(zip(productos,conteo_productos4.values()))

def top_porducto(diccionario_de_tienda,top_tienda,num_tienda):
  mas_vendido = None
  menos_vendido = None
  for llave,valor in diccionario_de_tienda.items():
    if valor==top_tienda[-1]:
      print(f"El Producto más vendido de Tienda {num_tienda} es: {llave}")
      mas_vendido=llave
    elif valor==top_tienda[0]:
      print(f"El Producto menos vendido de Tienda {num_tienda} es: {llave}")
      menos_vendido=llave
  return mas_vendido,menos_vendido

mas_vendido_T1, menos_vendido_T1 = top_porducto(dicP1, acomodoT1,1)
mas_vendido_T2, menos_vendido_T2 = top_porducto(dicP2, acomodoT2,2)
mas_vendido_T3, menos_vendido_T3 = top_porducto(dicP3, acomodoT3,3)
mas_vendido_T4, menos_vendido_T4 = top_porducto(dicP4, acomodoT4,4)
```
